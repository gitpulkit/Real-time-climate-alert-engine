```python
import requests
import json

# OpenWeatherMap API Key (Replace with your key)
API_KEY = "449cd6fbc4fd1e1e2a9e74f43b0689be"

# City for which we want weather data
CITY = "New York"

# API Endpoint
WEATHER_API_URL = f"https://api.openweathermap.org/data/2.5/weather?q={CITY}&appid={API_KEY}&units=metric"

def fetch_weather():
    """Fetch weather data from OpenWeatherMap API and print it."""
    try:
        response = requests.get(WEATHER_API_URL)
        data = response.json()

        if response.status_code == 200:
            # Extract relevant data
            print("\n✅ Weather Data Fetched Successfully:")
            print(f"City: {data['name']}")
            print(f"Temperature: {data['main']['temp']}°C")
            print(f"Humidity: {data['main']['humidity']}%")
            print(f"Weather: {data['weather'][0]['description'].capitalize()}")
            print(f"Wind Speed: {data['wind']['speed']} m/s")
            print(f"Timestamp: {data['dt']}")
        else:
            print(f"\n❌ Error: {data['message']} (Code {response.status_code})")

    except Exception as e:
        print(f"\n❌ Error fetching weather data: {e}")

# Run the function
fetch_weather()
```
